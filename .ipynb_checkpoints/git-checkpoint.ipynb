是什么 基本命令 区域划分 分支 git与github gitflow工作流 给开源项目贡献源代码

- git
    - 是什么
    
        分布式版本控制的软件
        
        - 理解
        
            - 分布式：与集中式相对应，集中式是有一个公共服务器，东西都存放在这个中央服务器，每个用户本地都不保存。分布式则弱化了这个中央服务器的概念，弱化不等于不重要哈，仍然和集中式的相同重要，分布式在每一个用户的本机那里都有自己的备份，也可以认为每台机器都可以是中央服务器。
            
       - 版本控制
           - 主要是应对新上线的版本因为一些原因必须得会滚到特定的版本。
            
    - 基本命令
        - init
            我的理解是给予git这个目录的操作权限

        - status
            查看当前文件的状态，是处于工作区还是缓存区

        - add
            将处于文件从工作区的推到缓存区

        - commit -m "version name"
            在本机生成版本
        
        - reset --hard 版本编码
            将主分支设定到特定版本
        
        - log/reflog --graph
            查看版本详细
            - 区别
                - log只能显示当前主分支以前的版本
                
                - reflog可以显示所有的版本
                
                - --graph 以可视化的方式显示版本之间的联系，和那个tree有点像
        
        - branch 
            - 查看分支
        
        - branch xxx  创建
            - -d 删除分支
        
        - checkout master/xxx 
            - 切换分支
        
        - merge 合并分支
            - 注意和合并到哪儿就切换到那个分支
            
        - rebase(变基)
            - 我的理解他就是做封装，就是将多个提交的几路车封装成一个
            
            - 使得提交记录更加简洁
        
        
    - 区域划分
    
        - 工作区 
            - 所在文件夹初始化后并且未add前的
        
        - 暂存区
            - add后的
        
        - 版本库
            - commit后的
            
        - 划分这个我觉得主要是留考虑的余地，
            
            
    - 分支
        - 做代码的隔离
        
        - 是什么
            保留共同的，记录增量
            
        - 问题紧急修复
            - 比如主分支出现问题了，赶紧创建分支，修复问题，merge即可
        
        
        
        
    - git与github
        - 本地向github仓库推代码
        
            - git remote add origin url 起个昵称
            
            - git push -u origin 分支
            
        - 从github仓库克隆代码
        
            - git clone url
            
        - 只拉下来更新的
            
            - git pull origin 分支
        
            
            
    - gitflow工作流
        
       - 在这我分两个分支，一个是正式版本，一个是开发版本
       
       - 在开发版本上每个工作成员创建自己的分支，一步步写好功能之后，经过测试，经过代码review，再merge到正式版本
       
            
    - 给开源项目贡献源代码
            
      - 找到开源项目
      
      - fork
          - 其功能是将项目源代码copy一份放到自己的仓库，应该是直接copy，不会像百度网盘那样所有文件只有一份，其他都是对这个共有文件的引用。
          
      - pull到自己的电脑
      
          - 优化代码

      - 向开源项目原作者发起pull requests
